# Activity: Find files with Linux commands
In this lab activity, you’ll navigate a Linux file structure, locate files, and read the contents of files. 

As a security analyst, it’s key that you know how to navigate, manage, and analyze files remotely via a Linux shell without a graphical user interface (GUI).
<hr>

**Tip:** 
* If the file name is too long, type the first few letters and press **TAB**
* `Alt` + `Shift` + down arrow = copy and paste (faster)
* `cd /` : navigate to home directory
* `grep -r "abc" .` : to search recursively the current folder   
* `grep -r "abc" /home/analyst/logs` : to search recursively in the path specified and also its sub-directories

<br>

### **Task 1. Get the current directory information**
In this task, you must use the commands you learned about to check the current working directory and list its contents.

* `pwd` : Display your working directory
  
* `ls` : Display the names of the files and directories in the current working directory.

<br>

### **Task 2. Change directory and list the subdirectories**

* `cd` : Change direction to subdirectory so for instance, from `/home/analyst` I want to go into `reports` directory,
  * type `cd reports`

    * Note please `ls` to see what are the subdirectories under it
  
* `ls` : Display the files and subdirectories in the `/home/analyst/reports` directory

<br>

### **Task 3. Locate and read the contents of a file**

* Type `cd users` : Navigate to the `/home/analyst/reports/users` directory.
  
* Type `ls` : List the files in the current directory.
* Type `cat Q1_added_users.txt` : Display the contents of the `Q1_added_users.txt` file.

<br>

### **Task 4. Navigate to a directory and locate a file**

* Type `cd ..` : Navigate back by one level to `/home/analyst/` directory.
  
* Type `ls` : List the files in the current directory.
* Type `cat server_logs.txt` : Display the contents of the `server_logs.txt` file.
* Type `head server_logs.txt` : Display the default first 10 lines of the file.
