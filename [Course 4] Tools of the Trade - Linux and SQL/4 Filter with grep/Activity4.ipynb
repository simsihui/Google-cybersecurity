# Activity: Filter with grep

### Introduction

In this lab, you’ll learn how to use the `grep` command and **piping** = `|` to search for files and return specific information. You’ll get information from different files, including server log files and user data files. You’ll use Linux commands in the Bash shell to complete these steps.

**What you’ll do:**
1. Search for error messages in a file
2. Search for files that contain a specific string
3. Search for information in user files

### **Task 1. Search for error messages in a log file**
navigate to the `/home/analyst/logs` directory and report on the error messages in the `server_logs.txt` file.

*  `cd logs` to navigate to the `/home/analyst/logs` directory
   * My ans : `cd /home/analyst/logs`
*  `grep error server_logs.txt`: to sieve through the entire data/text to return "error"
   *  My ans : `cat server_log.txt | grep error `
   * My ans : `cat server_log.txt | grep error | wc -l `: to return no. of documents containing "error"
  
Note: `ls /home/analyst/logs | grep server`: helps to sieve through all the files inside `logs` folder to get all documents with the word "server"
